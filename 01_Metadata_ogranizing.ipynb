```python
import numpy as np
import glob
import pandas as pd
```

```python
names=glob.glob("Training_WFDB/*.hea")
names=np.array(names)
```

```python
for i in range(names.shape[0]):
    names[i]=names[i][14:23]
    print(names[i])
```

```python
Age=np.copy(names)
Sex=np.copy(names)
Dx=np.copy(names)
Dx_first=np.copy(names)
Dx_second=np.copy(names)
Rx=np.copy(names)
Hx=np.copy(names)
Sx=np.copy(names)
Names=np.copy(names)
```

```python
for i in range(names.shape[0]):
    file=open('Training_WFDB/'+names[i])
    lines = file.readlines()[1:]
    Age[i]=lines[12].split()[1]
    Sex[i]=lines[13].split()[1]
    Dx[i]=lines[14].split()[1]
    Dx_second[i]=('-')
    Rx[i]=lines[15].split()[1]
    Hx[i]=lines[16].split()[1]
    Sx[i]=lines[17].split()[1]
    Names[i]=lines[0].split()[0][0:5]
```

```python
for i in range(names.shape[0]):
    Dx_first[i]=Dx[i].split(",")[0]
    if (len(Dx[i].split(","))==2):
        Dx_second[i]=Dx[i].split(",")[1]
```

```python
df = pd.DataFrame(data={'Names': Names, 'Age': Age, 'Sex': Sex, 'Dx': Dx, 'Dx_first': Dx_first, 'Dx_second': Dx_second, 'Rx': Rx, 'Hx': Hx, 'Sx': Sx}, index=names)
```

```python
df.to_csv(r'Metadata.csv', index = True)
```
